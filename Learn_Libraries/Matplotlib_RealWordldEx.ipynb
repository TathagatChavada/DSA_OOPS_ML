Gas Prices across the countries 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

gas = pd.read_csv(r'gas_prices.csv')
plt.figure(figsize=(11,7))


plt.title("Comparison of Gas Prices over the different years around the world")
plt.xlabel("Years")
plt.ylabel("Gas price (in US $)")

plt.xticks(gas.Year)

plt.plot(gas['Year'], gas['USA'],'o-', label ="USA")
plt.plot(gas['Year'], gas['Canada'],'o-', label ="Canada")

# Another way of plotting countries
# for country in gas:
#     if country != "Year":
#         plt.plot(gas.Year, gas[country],label = country ,marker= 'o')


plt.legend()

plt.show()
```

FIFA Player's Stats

```python
fifa = pd.read_csv(r'fifa_data.csv')
fifa.head(3)
```

Histogram

```python
bin = np.arange(40,101,10)

plt.title("Analysis of Overall Scores of different players")
plt.xlabel("Overall Score")
plt.ylabel("No. of Players")

plt.xticks(bin)

plt.hist(fifa.Overall, bins= bin)  


plt.show()
```

Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == "Left"].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == "Right"].count()[0] 

foot = ["Left", "Right"]
color = ['#845EC2','#FF6F91']

plt.title("Foot preference of fifa players")

plt.pie([left,right], labels = foot, shadow=True, colors=color, autopct='%.2f%%')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]


light = fifa[(fifa.Weight < 125)].count()[0]

light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]

medium = fifa[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]

medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight <200)].count()[0]

heavy = fifa[(fifa.Weight > 200)].count()[0]


weights = [light, light_medium, medium, medium_heavy, heavy]
weight_category = ['Under 125', '125-150', '150-175','175-200','Above 200']
myexp = [0.8,0,0,0,0.4]

plt.title("Weight Distribution of Players (in lbs)")
plt.pie(weights,labels=weight_category,autopct='%.2f%%', explode=myexp)
plt.show()
```

Box & Wishkers Chart

```python
barcelona = fifa.loc[(fifa.Club == "FC Barcelona")]['Overall']
madrid = fifa.loc[(fifa.Club == "Real Madrid")]['Overall']
england = fifa.loc[(fifa.Club == "New England Revolution")]['Overall']

clubs= [barcelona, madrid, england]
label = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

plt.title("Soccer Team Comparison")
plt.xlabel("FIFA Teams")
plt.ylabel("FIFA Overall Rating")

plt.boxplot(clubs, labels=label)
plt.show()
```
